Model kurmak, bağımlı değişken ile bağımsız değişkenler arasındaki ilişkiyi modellemek demektir. (Amacımız bağımlı değişken ile bağımsız değişkenler arasındaki ilişkiyi modellemeye çalışmak)

Burada modellerin başarısını değerlendirmek için daha doğru modellendirme şansı imkanı veren yöntemleri inceliyoruz.

# Model Doğrulama (Tuning) Yöntemleri
- Holdout
- K-katlı Çaprazlama
- Leave One Out
- Bootstrap

    Holdout Yöntemi:

Orijinal veri setini eğitim seti ve test seti olmak üzere ikiye ayırıyoruz. Daha sonrasında eğitim setiyle birlikte algoritmamızı eğitiyoruz. En sonda ise test seti ile birlikte eğittiğimiz algoritmanın doğru sonuç üretip üretmediğini test ediyoruz.



    K-Katlı Çapraz Sorgulama !!!! (K Fold Cross Validation): 

Öncelikle eğitim setimiz k tane parçaya ayırlır (genelde 5 ya da 10). Daha sonra ayrılan parçalardan bir tanesi dışarıda bırıkılır diğerleri ile model oluşturulur. Daha sonra dışarıda bırakılan parça ile model test edilir. Ve bu işlem k tane ayrılan parça için tekrar edilir. Bu işlemler sonucunda elde edilen hataların ortalaması alındığında bir "validasyon (doğrulama) hata oranı" elde ederiz. Daha sonra çalışmanın en başında bölmüş olduğumuz test setiyle birlikte de modelimizi son kez test ederiz.



    Leave One Out:
    
K Katlı Çapraz Doğrulama yönteminin özel bir halidir. Fark olarak buradaki k sayısı (yani kaç parçaya ayıracağımız) veri setindeki örnek sayısı olan n'ye eşittir. Yani eğitim setimizi örnek sayısı kadar parçaya böleriz. 



    Bootsrap Yöntemi:

Elimizdeki orijinal veri setinin içerisinden orijinal veri setinin içerisindeki gözlem sayısından daha az sayıda gözlem içeren bootsrap örneklemleri oluşturulur. Bu örneklemler üzerinden model kurulur. Kurulmuş olan modeller test seti yaklaşımıyla test edilip sonuçları karşılaştırılarak değerlendirilir. Netice olarak bu yöntem yerine koymalı bir şekilde veri seti içerisinden veri türetmek şeklinde kullanılır ve oluşan yeni verilerin her biri üzerinden model kurulur, test edilir, değerlendirilir.
