Modelin başarısını değerlendirirken elimizde iki tip hata olur:

- Eğitim Hatası -> Modeli kurmak için kullanılan veri seti üzerinden elde edilen hatadır. Eğitim hatası test hatasının yanlı ve kötü bir tahmincisidir. 

- Test Hatası -> Kurulan modeli test etmek için kullandığımız veri seti için oluşan hatadır.

Esneklik, verinin fonksiyonel yapısının uygun bir şekilde yorumlanmasıdır. (esneklik = verinin yapısına kendisini uydurma. Aşağıdakilerden en esneği en sağdaki)

Yanlılık (Underfitting), gerçek değerler ile tahmin edilen değerler arasındaki farkı ifade eder. (ilk görsel yanlıdır. Gerçek verileri başarılı tahmin edememiş -az öğrenmiş-)

Varyans = esneklik. Varyans yüksekse esneklik yüksektir ama başka bir veri seti için uygulandığında başarılı olmaz çünkü ezber yapmıştı.



    Eğitim Hatası ile Test Hatasının Karşılaştırılması:

Bu grafiği inceleyerek modeli eğitirken nerede durmamız gerektiğine karar vermeye çalışırız. Grafikte görüldüğü üzere eğitim setinde hata oranı düşüyor ama test setinde düştükten bir müddet sonra tekrar artmaya başlıyor. İşte tam bu noktada modeli eğitmeyi durdurmamız gerekiyor.

Özetle, verimizi ikiye ayırdığımızda eğitim için bir set , test etmek için de bir setimiz var. Modeli eğitmek için kullanmış olduğumuz eğitim setini öyle bir kullanmalıyız ki genellenebilirlik kaygımızdan uzaklaşmayalım. Eğitim  setini öğrenelim ama ezberlemeyelim.
