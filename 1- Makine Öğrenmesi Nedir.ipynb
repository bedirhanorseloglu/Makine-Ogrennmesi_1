Bilgisayarların insanlara benzer şekilde öğrenmesini sağlamak amacıyla çeşitli algoritma ve tekniklerin geliştirilmesi için çalışılan bilimsel çalışma alanıdır.



    Temel Kavramlar:

- Bağımlı (Hedef, Çıktı, Yanıt) Değişken -> İlgilenmiş olduğumuz makine öğrenmesi probleminde hedeflediğimiz, odaklandığımız ana değişkenimiz (araç fiyatı -> y1)

- Bağımsız Değişken -> Bağımlı değişkeni oluşturduğu varsayılan değişkenlerdir. (vites türü, km, hasar durumu, marka, model -> x1 , x2 , x3 , x4)

- Gözetimli Öğrenme -> Bağımlı değişken ve bağımlı değişkeni meydana getiren bağımsız değişkenler bir aradaysa bu duruma gözetimli öğrenme denir. 



- Gözetimsiz Öğrenme -> Bağımlı değişkenin çalışmanın içinde olmadığı öğrenme türüdür. Amacımız gözlem birimlerini birbirine benzer özelliklerine göre bir araya getirmektir.



- Regresyon Problemi -> Bağımlı değişken sayısal/sürekli bir değişken ise bu bir regresyon problemidir.

- Sınıflandırma Problemi -> Bağımlı değişken kategorik bir değişken ise bu sınıflandırma problemidir.

- Test-Train -> Elimizdeki orijinal veri setini eğitim seti ve test seti olmak üzere ikiye ayırırız.



- Değişken Seçimi -> Bağımlı değişkene ulaşabilmek için seçmemiz gereken bağımsız değişkenlerdir. (Amacımız en az değişenle en fazla açıklanabilirliği yakalamaya çalışmak)

- Model Seçimi -> İlk olarak oluşabilecek değişken kombinasyonları ile oluşturulan modeller arasında en iyi modelin seçilmesi. İkincisi, kurulan birbirinden farklı modeller arasından model seçimi.       
Peki model neye göre seçilir?:
    - Regresyon için açıklanabilirlik oranı ve RMSE (bir hata ölçüm metriği -> Root - Mean - Square Deviation) türevi bir değer
    - Sınıflandırma için doğru sınıflandırma oranı türevi bir değer

- Aşırı Öğrenme (Overfitting) -> Algoritmanın kendisine verilen veri setinin yapısını çok iyi öğrenip (ezberleyip) görmemiş olduğu yeni veriler üzerinden tahmin yapmak istendiğinde başarısız olması

- Deterministik (Kesin) Modeller -> Değişkenler arasında kesin bir ilişki olduğunu varsayan modellerdir. (örneğin y = 2x doğrusu gibi. Noktalar doğrununun hemen yanında)



- Stokastik (Olasılıksal) Modeller -> Değiikenler arasında tesadüfi bir hata oranıyla ilişki olduğunu varsayan model (örneğin y = 2x + tesadüfi hata. Noktalar doğrunun yakınlarında)
