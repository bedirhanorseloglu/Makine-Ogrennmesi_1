Modellerin tahmin başarılarını değerlendireceğimiz yöntemleri inceleyeceğiz. Bir önceki konuyla arasındaki farkı anlamamız önemli.

Regresyon ve sınıflandırma problemleri için başarı değerlendirme yöntemlerini inceliyoruz.

# Regresyon Modelleri İçin Başarı Değerlendirme Yöntemleri
- MSE
- RMSE
- MAE

# Sınıflandırma Modelleri İçin Başarı Değerlendirme Yöntemleri
- Tablo
- ROC Eğrisi


    Regresyon Modelleri İçin Başarı Değerlendirme:
    
- İlk başarı değerlendirme kriterimiz MSE. Biz bir model kurduğumuzda bu modelin amacı bağımlı değişkenin değerlerini tahmin etmek olacak. Ama bizim elimizde bağımlı değişkenin gerçek değerleri ve tahmin ettiğimiz değerler zaten var. Dolayısıyla bize günün sonunda ortalama ne kadar hata yaptığımızı yani modelimizin başarısını değerlendirmek istediğimizde bu formül işimize yarıyor.






- Bir diğer başarı değerlendirme kriterimiz RMSE. Aslında burada yaptığımız işlem bir üstte bahsettiğimiz değerin karekökünü almak. 



- Bir diğer başarı değerlendirme kriteri MAE (Ortalama Mutlak Hata). 



    Sınıflandırma Modelleri İçin Başarı Değerlendirme:




- Bir diğer kriterimiz ROC Eğrisi. Bu eğrinin altında kalan alana AUC (Area Under Curve - Eğri Altındaki Alan) denir. Bu alanın büyük olması modelin başarılı olduğu , küçük olması başarısız olduğu anlamına gelir. Mavi çizgi ortadaki kesikli çizgiye ne kadar yaklaşıyorsa modelin başarısı o kadar düşüyor demektir. Ortadaki çizgiler ise hiç model kurmasaydık ve bütün sınıflara 1 veya 0 deseydik yani rastgele bu işlemi yapsaydık elde edeceğimiz başarıyı (%50) ifade ediyor.
